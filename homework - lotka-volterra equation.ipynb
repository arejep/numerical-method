lotka-volterra equation
======
##### Lotka-Volterra equations describe the interactions between predator and prey populations in ecosystem
 
dN/dt = r*N - a*N*P

dP/dt = -sP +bNP
- r is the intrinsic growth rate of the prey
- a is the rate at which predators kill prey
- s is the death rate of predators in the absence of prey
- b is the rate at which succesful predation increases predator numbers

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Lotka-Volterra equations
def lotka_volterra(y, t, r, a, s, b):
    N, P = y
    dNdt = r * N - a * N * P
    dPdt = -s * P + b * N * P
    return [dNdt, dPdt]

# Parameters
r = 0.3   # prey intrinsic growth rate
a = 0.03  # rate at which predators kill prey
s = 0.2   # predator death rate in the absence of prey
b = 0.02  # rate at which successful predation increases predator numbers

# Initial populations
N0 = 30  # initial prey population
P0 = 8   # initial predator population
initial_conditions = [N0, P0]

# Time points
t = np.linspace(0, 200, 1000)

# Differential equations using odeint
solution = odeint(lotka_volterra, initial_conditions, t, args=(r, a, s, b))

# Plot the results
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, solution[:, 0], label='Prey (N)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, solution[:, 1], label='Predator (P)', color='orange')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()

plt.tight_layout()
plt.show()


```
